```python
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]

# Import pandas as pd
import pandas as pd

# Create dictionary my_dict with three key:value pairs: my_dict
my_dict = {'country': names, 'drives_right': dr, 'cars_per_cap': cpc}

# Build a DataFrame cars from my_dict: cars
cars = pd.DataFrame(my_dict)

# Print cars
print(cars)
```

```python
rowlabels= ["US", "AU", "J", "IN", "RU", "MO", "EGY"]
cars.index = rowlabels
print(cars)
```

```python
print(cars[['country']])
type(cars[['country']])
#double square brackets result in DataFrame type of output
```

```python
#access columns using square brackets
cars[['country', 'drives_right']]
```

```python
#access rows using square brackets- only through slicing
cars[0:4]
```

```python
#row access using loc
#loc is label- based
cars.loc[['US','J','EGY'], :]
```

```python
#access columns using loc
cars.loc[:, ['country', 'cars_per_cap']]
```

```python
#access both rows and columns using loc
cars.loc[['RU', 'MO', 'J'], ['country', 'drives_right']]
```

```python
#row access using iloc
#iloc is integer index based
cars.iloc[[0, 1, 2, 3]]
```

```python
#column access using iloc
cars.iloc[:, [0, 2]]
```

```python
#access both rows and columns using iloc
print(cars.iloc[[1, 2, 3], [0, 1]])
```
