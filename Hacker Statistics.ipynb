```python
import numpy as np

np.random.seed(123)
print(np.random.rand())
```

```python
#rolling a dice to determine my next move
step = 50
dice = np.random.randint(1,7)

if dice <= 2 :
    step -= 1
elif dice > 2 and dice < 6 :
    step += 1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print(dice)
print(step)
```

```python
#generate a random walk from a random step

random_walk = [0]

for x in range(100) :
    # step is set to the last element of the random_walk
    step = random_walk[-1]

    dice = np.random.randint(1,7)

    if dice <= 2:
        step = max(0, step - 1) #to ensure that the step does not go below 0
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

print(random_walk)

import matplotlib.pyplot as plt
plt.plot(random_walk)
plt.show()
```

```python
all_walks = []

# Simulate random walk 10 times
for i in range(500) :

    # Code from before
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
        
        # Implement clumsiness, the probability of falling, step is reset to 0
        if np.random.rand() <= 0.001 :
            step = 0

    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
#print(all_walks)
```

```python
# Convert all_walks to Numpy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()
# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

```python
ends = np_aw_t[-1, :]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```
