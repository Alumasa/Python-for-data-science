```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
read_file = pd.read_csv(r'datasets/Grants 2010.txt')
read_file.to_csv(r'datasets/Grants 2010.csv', index=None)
```

```python
df = pd.read_csv('datasets/Grants 2010.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df['Region'].fillna(df['Region'].describe().top, inplace=True)
```

```python
# Display pandas histogram
df['Award_Amount'].plot.hist()
plt.show()

# Clear out the pandas histogram
plt.clf()
```

```python
# Display a Seaborn distplot
sns.distplot(df['Award_Amount'])
plt.show()

# Clear the distplot
plt.clf()
```

```python
# Create a distplot
sns.distplot(df['Award_Amount'],
             kde=False,
             bins=20)

# Display the plot
plt.show()
```

```python
# Create a distplot
sns.distplot(df['Award_Amount'],
             kde=True,
             bins=20)

# Display the plot
plt.show()
```

```python
# Create a distplot of the Award Amount
sns.distplot(df['Award_Amount'],
             hist=False,
             rug=True,
             kde_kws={'shade':True})

# Plot the results
plt.show()
```

```python
read_file = pd.read_csv(r'datasets/Automobile Insurance.txt')
read_file.to_csv(r'datasets/Automobile Insurance.csv', index=None)
```

```python
insurance = pd.read_csv('datasets/Automobile Insurance.csv')
insurance.head()
```

```python
# Create a regression plot of premiums vs. insurance_losses
sns.regplot(data=insurance, x='insurance_losses', y='premiums')



# Display the plot
plt.show()
```

```python
# Create an lmplot of premiums vs. insurance_losses
sns.lmplot(x='insurance_losses', y='premiums', data=insurance)



# Display the second plot
plt.show()
```

```python
# Create a regression plot using hue
sns.lmplot(data=insurance,
           x="insurance_losses",
           y="premiums",
           hue="Region")

# Show the results
plt.show()
```

```python
# Create a regression plot with multiple rows
sns.lmplot(data=insurance,
           x="insurance_losses",
           y="premiums",
           col="Region")

# Show the plot
plt.show()
```
