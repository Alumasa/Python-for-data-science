### Color with Seaborn

```python
# Set style, enable color code, and create a magenta distplot
sns.set(color_codes=True)
sns.distplot(data['fmr_3'], color='m')

# Show the plot
plt.show()
```

```python
# Loop through differences between bright and colorblind palettes
for p in ['bright', 'colorblind']:
    sns.set_palette(p)
    sns.distplot(data['fmr_3'])
    plt.show()
    
    # Clear the plots    
    plt.clf()
```

```python
#sequential colors: consistent range of data from high to low
sns.palplot(sns.color_palette('Purples', 8))
plt.show()
```

```python
#circular colors: unordered data
sns.palplot(sns.color_palette('husl', 10))
plt.show()
```

```python
#diverging colors: both  low and high values are interesting
sns.palplot(sns.color_palette('coolwarm', 6))
plt.show()
```

```python
#diverging
sns.palplot(sns.color_palette('BrBG', 12))
plt.show()
```

### Axes labels

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the distribution of data
sns.distplot(data['fmr_3'], ax=ax)

# Create a more descriptive x axis label
ax.set(xlabel="3 Bedroom Fair Market Rent")

# Show the plot
plt.show()
```

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the distribution of 1 bedroom rents
sns.distplot(data['fmr_1'], ax=ax)

# Modify the properties of the plot
ax.set(xlabel="1 Bedroom Fair Market Rent",
       xlim=(100,1500),
       title="US Rent")

# Display the plot
plt.show()
```

```python
# Create a figure and axes. Then plot the data
fig, ax = plt.subplots()
sns.distplot(data['fmr_1'], ax=ax)

# Customize the labels and limits
ax.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500), title="US Rent")

# Add vertical lines for the median and mean
ax.axvline(x=data['fmr_1'].median(), color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=data['fmr_1'].mean(), color='b', label='Mean', linestyle='-', linewidth=2)

# Show the legend and plot the data
ax.legend()
plt.show()
```

```python
# Create a plot with 1 row and 2 columns that share the y axis label
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True)

# Plot the distribution of 1 bedroom apartments on ax0
sns.distplot(data['fmr_1'], ax=ax0)
ax0.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500))

# Plot the distribution of 2 bedroom apartments on ax1
sns.distplot(data['fmr_2'], ax=ax1)
ax1.set(xlabel="2 Bedroom Fair Market Rent", xlim=(100,1500))

# Display the plot
plt.show()
```

### Categorical Plot Types

#### stripplot() and swarmplot()

```python
import pandas as pd
df = pd.read_csv('datasets/Grants 2010.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create the stripplot
sns.stripplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         jitter=True)

plt.show()
```

```python
# Create and display a swarmplot with hue set to the Region
sns.swarmplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         hue='Region')

plt.show()
```

#### boxplots, violinplots and lvplots

```python
# Create a boxplot
sns.boxplot(data=df,
         x='Award_Amount',
         y='Model Selected')

plt.show()
plt.clf()
```

```python
# Create a violinplot with the husl palette
sns.violinplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='husl')

plt.show()
plt.clf()
```

```python
# Create a lvplot with the Paired palette and the Region column as the hue
sns.boxenplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='Paired',
         hue='Region')

plt.show()
plt.clf()
```

#### barplots, pointplots and countplots

```python
# Show a countplot with the number of models used with each region a different color
sns.countplot(data=df,
         y="Model Selected",
         hue='Region')

plt.show()
plt.clf()
```

```python
# Create a pointplot and include the capsize in order to show bars on the confidence interval
sns.pointplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         capsize=.1)

plt.show()
plt.clf()
```

```python
# Create a barplot with each Region shown as a different color
sns.barplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         hue='Region')

plt.show()
plt.clf()
```

### Regression plots

```python
read_file = pd.read_csv(r'datasets/College Tuition 2018.txt')
read_file.to_csv(r'datasets/College Tuition 2018.csv', index=None)
```

```python
read_file.head()
```

```python
df = pd.read_csv('datasets/College Tuition 2018.csv')
df.head()
```

```python
# Display a regression plot for Tuition
sns.regplot(data=df,
         y='Tuition',
         x="SAT_AVG_ALL",
         marker='^',
         color='g')

plt.show()
plt.clf()
```

```python
# Display the residual plot
sns.residplot(data=df,
          y='Tuition',
          x="SAT_AVG_ALL",
          color='g')

plt.show()
plt.clf()
```

```python
# Plot a regression plot of Tuition and the Percentage of Pell Grants
sns.regplot(data=df,
            y='Tuition',
            x="PCTPELL")

plt.show()
plt.clf()
```

```python
# Create another plot that estimates the tuition by PCTPELL
sns.regplot(data=df,
            y='Tuition',
            x="PCTPELL",
            x_bins=5)

plt.show()
plt.clf()
```

```python
# The final plot should include a line using a 2nd order polynomial
sns.regplot(data=df,
            y='Tuition',
            x="PCTPELL",
            x_bins=5,
            order=2)

plt.show()
plt.clf()
```

```python
# Create a scatter plot by disabling the regression line
sns.regplot(data=df,
            y='Tuition',
            x="UG",
            fit_reg=False)
```

```python
# Create a scatter plot and bin the data into 5 bins
sns.regplot(data=df,
            y='Tuition',
            x="UG",
            x_bins=5)

plt.show()
plt.clf()
```

```python
# Create a regplot and bin the data into 8 bins
sns.regplot(data=df,
         y='Tuition',
         x="UG",
         x_bins=8)

plt.show()
plt.clf()
```
